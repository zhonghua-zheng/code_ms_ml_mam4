```python
import numpy as np
import pandas as pd
import xarray as xr
```

```python
year = "2011"
file_path = {}
file_path["MAM4"] = "/data/keeling/a/zzheng25/d/mam4_paper_data/chi_only/mam4_chi/"
file_path["ML"] = "/data/keeling/a/zzheng25/d/mam4_paper_data/chi_only/ml_chi/"
file_path["diff"] = "/data/keeling/a/zzheng25/d/mam4_paper_data/chi_only/mam4_minus_ml_chi/"
file_path["diff_abs"] = "/data/keeling/a/zzheng25/d/mam4_paper_data/chi_only/mam4_minus_ml_chi/"
mask_path = "/data/keeling/a/zzheng25/d/mam4_paper_data/chi_only/mask/"
comp_path = "/data/keeling/a/zzheng25/d/mam4_paper_data/chi_only/comp_analysis/"
```

```python
chi = "chi_b"
mask = xr.open_dataset(mask_path+str(year)+"_"+chi+".nc")["mask"]
ds_comp = xr.open_dataset(comp_path+str(year)+"_year_comp.nc").where(mask)
ds_MAM4 = xr.open_dataset(file_path["MAM4"]+str(year)+"_"+chi+"_mean.nc").where(mask)
ds_ML = xr.open_dataset(file_path["ML"]+str(year)+"_"+chi+"_mean.nc").where(mask)
```

```python
df_comp = ds_comp.to_dataframe().reset_index()
df_MAM4 = ds_MAM4.to_dataframe().reset_index()
df_ML = ds_ML.to_dataframe().reset_index()
df_comp["bc_per"] = df_comp["bc_a1_per"]+df_comp["bc_a4_per"]
df_comp["bc_SRF"] = df_comp["bc_a1_SRF"]+df_comp["bc_a4_SRF"]
```

```python
df_comp_MAM4 = pd.merge(df_comp, df_MAM4, on=["lat","lon"],how="inner")
df_comp_ML = pd.merge(df_comp, df_ML, on=["lat","lon"],how="inner")
```

```python
print("MAM4 overall")
display(df_comp_MAM4.describe()[["bc_per",chi]].transpose())

print("MAM4 high bc SRF")
display(df_comp_MAM4[df_comp_MAM4["bc_SRF"]>df_comp_MAM4["bc_SRF"]\
                     .quantile(0.95)].describe()[["bc_per",chi]].transpose())

print("MAM4 high bc per")
display(df_comp_MAM4[df_comp_MAM4["bc_per"]>df_comp_MAM4["bc_per"]\
                     .quantile(0.95)].describe()[["bc_per",chi]].transpose())
```

```python
print("ML overall")
display(df_comp_ML.describe()[["bc_per",chi]].transpose())

print("ML high bc SRF")
display(df_comp_ML[df_comp_ML["bc_SRF"]>df_comp_ML["bc_SRF"]\
                     .quantile(0.95)].describe()[["bc_per",chi]].transpose())

print("ML high bc")
display(df_comp_ML[df_comp_ML["bc_per"]>df_comp_ML["bc_per"]\
                   .quantile(0.95)].describe()[["bc_per",chi]].transpose())
```
